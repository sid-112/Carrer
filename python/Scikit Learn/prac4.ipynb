```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('train_u6lujuX_CVtuZ9i (1).csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df['Loan_ID'] = df['Loan_ID'].str.replace('LP00', '')
df['Loan_ID'] = df['Loan_ID'].astype(int)
```

```python
df['Gender'] = df['Gender'].replace(to_replace='Male', value=1)
df['Gender'] = df['Gender'].replace(to_replace='Female', value=0)

df['Gender'] = df['Gender'].astype(int)
```

```python
df['Married'] = df['Married'].replace(to_replace='Yes', value=1)
df['Married'] = df['Married'].replace(to_replace='No', value=0)

df['Married'] = df['Married'].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Dependents'] = df['Dependents'].str.replace('+', '')
df['Dependents'] = df['Dependents'].astype(int)
```

```python
df['Dependents'].unique()
```

```python
df.info()
```

```python
df.head()
```

```python
df['Education'] = df['Education'].replace(to_replace='Graduate', value=1)
df['Education'] = df['Education'].replace(to_replace='Not Graduate', value=0)

df['Education'] = df['Education'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Self_Employed'] = df['Gender'].replace(to_replace='Yes', value=1)
df['Self_Employed'] = df['Gender'].replace(to_replace='No', value=0)

df['Self_Employed'] = df['Gender'].astype(int)
```

```python
df.info()
```

```python
df['Property_Area'].unique()
```

```python
df.head()
```

```python
df['Property_Area'] = df['Property_Area'].replace(to_replace='Rural', value=1)
df['Property_Area'] = df['Property_Area'].replace(to_replace='Urban', value=0)
df['Property_Area'] = df['Property_Area'].replace(to_replace='Semiurban', value=2)


df['Property_Area'] = df['Property_Area'].astype(int)
```

```python
df.info()
df.head()
```

```python
df['Loan_Status'] = df['Loan_Status'].replace(to_replace='Y', value=1)
df['Loan_Status'] = df['Loan_Status'].replace(to_replace='N', value=0)

df['Loan_Status'] = df['Loan_Status'].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop(columns=['Loan_Status'])
y =  df['Loan_Status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
step1 = StandardScaler()

step2 = svm.SVC(kernel='linear')
```

```python
pipe = make_pipeline(step1, step2)
```

```python
pipe
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
print(mean_squared_error(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred3 = pipe.predict(X_test)

print(mean_squared_error(y_test, y_pred3))
print(accuracy_score(y_test, y_pred3))
```
