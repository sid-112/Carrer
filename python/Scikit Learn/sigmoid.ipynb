```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
  """Calculates the sigmoid function for a given input x."""
  return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 100)  # Create 100 evenly spaced points from -10 to 10

# Calculate y values (sigmoid of x)
y = sigmoid(x)

# Plot the sigmoid function
plt.figure(figsize=(8, 6))  # Adjust the figure size
plt.plot(x, y, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True) # adds a grid
plt.axhline(0, color='black',linewidth=0.5) #add horizontal line at y=0
plt.axhline(1, color='black',linewidth=0.5) #add horizontal line at y=1
plt.axvline(0, color='black',linewidth=0.5) #add vertical line at x=0
plt.legend()
plt.show()
```
