```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with 'studytime' and 'score' columns
# Replace this with your actual DataFrame
# Example:
data = {'studytime': np.random.rand(100) * 100, 'score': np.random.rand(100) * 100} # Example data
df = pd.DataFrame(data)

# Extract data
X = df['studytime']
Y = df['score']

# Initialize parameters
m = 0
b = 0

# Number of data points
n = 100

# Learning rate
alpha = 0.0001  # Adjust as needed

# Number of iterations
iterations = 1000  # Adjust as needed

# Gradient Descent Loop
for i in range(iterations):
    y = m * X + b
    A = (-2/n) * np.sum(m * X.ravel() + b - y.ravel() * X.ravel() + 2 * alpha * m)
    B = (-2/n) * np.sum(m * X.ravel() + b - y.ravel())
    m = m - alpha * A
    b = b - alpha * B

# Plotting
y_pred = m * X + b
plt.scatter(X, Y, label='Original Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Study Time')
plt.ylabel('Score')
plt.title('Linear Regression')
plt.legend()
plt.show()

print(f"Final m: {m}")
print(f"Final b: {b}")
```
