```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge, LinearRegression, Lasso, LogisticRegression
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.shape
```

```python
df.head()
```

```python
df['score'] = (df['math score'] + df['reading score'] + df['writing score'])
```

```python
df = df.drop(columns=['math score', 'reading score', 'writing score'])
```

```python
df.head(7)
```

```python
X = df.drop(columns=['score'])
y = df['score']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
step1 = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(drop='first'), [0, 1, 2, 3, 4])
])

step2 = Lasso(alpha=0.01)
```

```python
pipe = make_pipeline(step1, step2)
```

```python
pipe
```

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

```

```python
print(r2_score(y_test, y_pred))
```

```python
step1 = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(drop='first'), [0, 1, 2, 3, 4])
])

step2 = KNeighborsRegressor(n_neighbors=50)

pipe = make_pipeline(step1, step2)
```

```python
pipe.fit(X_train, y_train)
y_pred2 = pipe.predict(X_test)
```

```python
print(r2_score(y_test, y_pred2))
```

```python
step1 = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(drop='first'), [0, 1, 2, 3, 4])
])

step2 =Ridge(alpha=1)

pipe = make_pipeline(step1, step2)
```

```python
pipe
```

```python
pipe.fit(X_train, y_train)
y_pred3 = pipe.predict(X_test)

print(r2_score(y_test, y_pred3))
```

```python
step1 = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(drop='first'), [0, 1, 2, 3, 4])
])

step2 =LinearRegression()

pipe = make_pipeline(step1, step2)

pipe.fit(X_train, y_train)
y_pred4 = pipe.predict(X_test)

print(r2_score(y_test, y_pred4))
```

```python
step1 = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(drop='first'), [0, 1, 2, 3, 4])
])

step2 =LogisticRegression()

pipe = make_pipeline(step1, step2)

pipe.fit(X_train, y_train)
y_pred5 = pipe.predict(X_test)

print(r2_score(y_test, y_pred5))
```

```python
step1 = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(drop='first'), [0, 1, 2, 3, 4])
])

step2 =KNeighborsClassifier()

pipe = make_pipeline(step1, step2)

pipe.fit(X_train, y_train)
y_pred6 = pipe.predict(X_test)

print(r2_score(y_test, y_pred6))
```
